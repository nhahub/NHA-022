```python
import geopandas as gpd
from keplergl import KeplerGl
from shapely.geometry import LineString, Polygon
```

```python
boundaries = gpd.read_file('../data/egypt/egy_admbndl_admALL_capmas_itos_20170421.shp').to_crs(epsg=4326)
alex = gpd.read_file('../data/egypt/alexandria.geojson').to_crs(epsg=4326).geometry.iloc[0]
```

```python
alex_inner_boundaries_mask = boundaries.within(alex)

alex_inner_boundaries = boundaries[alex_inner_boundaries_mask]

alex_inner_boundaries = alex_inner_boundaries['geometry']
```

```python
# Convert closed LineStrings to Polygons
def line_to_polygon(geom):
    if not geom.is_ring:
        # close it if it's not closed
        coords = list(geom.coords)
        if coords[0] != coords[-1]:
            coords.append(coords[0])
        geom = LineString(coords)
    return Polygon(geom)

alex_inner_boundaries = alex_inner_boundaries.apply(line_to_polygon)
```

```python
test = KeplerGl()
test.add_data(data=alex_inner_boundaries.reset_index())
test
```
